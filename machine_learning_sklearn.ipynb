Modelo con scikitlearn

```python
import pandas as pd
from textblob import TextBlob
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
df_reviews = pd.read_csv("ETL\datasets\df_reviews.csv")
```

```python
df_reviews
```

```python
df_reviews['sentiment'] = df_reviews['text'].apply(lambda x: TextBlob(str(x)).sentiment.polarity)
# Etiquetar como 'positivo' si el sentimiento es mayor que 0, 'negativo' si es menor que 0 y 'neutral' si es 0
df_reviews['sentiment_label'] = df_reviews['sentiment'].apply(lambda x: 'positivo' if x > 0 else 'negativo' if x < 0 else 'neutral')
```

```python
df_reviews = pd.read_csv("ETL\datasets\df_reviews.csv")
```

```python
df_reviews
```

```python
df_reviews['text'] = df_reviews['text'].astype(str)
```

```python
# Crear un objeto SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
```

```python
# Aplicar el análisis de sentimiento a cada fila en la columna 'text'
df_reviews['compound'] = df_reviews['text'].apply(lambda x: sia.polarity_scores(x)['compound'])
```

```python
# Asignar etiquetas de sentimiento basadas en el puntaje compuesto
df_reviews['sentiment'] = df_reviews['compound'].apply(lambda x: 'positivo' if x >= 0 else 'negativo')
```
